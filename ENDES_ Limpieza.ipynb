```python
import pandas as pd
import pyreadstat
import os

path= os.path.join('data','Endes_2022_N&M','786-Modulo1632' ,'RE223132.sav')
path3= os.path.join ('data','Endes_2022_MEF','786-Modulo1631','REC0111.sav')
data,endes = pyreadstat.read_sav(path)
data3, ende3 = pyreadstat.read_sav(path3)
```

```python
variables =['CASEID','V213','V301', 'V302', 'V393', 'V3A06']

data1=data[variables]
data1 =pd.DataFrame(data1)
data1
```

```python
var = ['CASEID', 'UBIGEO', 'V013' ]
data3 =data3[var]
data3=pd.DataFrame(data3)
data3
```

```python
endes2022= pd.merge(data1, data3, on='CASEID') 
```

```python
endes2022
```

```python
endes2022['IDPROV'] = endes2022['UBIGEO'].astype(str).str[:4]
endes2022['IDDPTO'] = endes2022['UBIGEO'].astype(str).str[:2]
```

```python
nuevos_nombres = {'V213': 'act_embara', 'V301': 'conoce_meto', 'V302': 'uso_meto', 'V393': 'visita_planfami', 'V3A04':'info_meto', 
                'V013': 'Edad'}
endes2022 = endes2022.rename(columns=nuevos_nombres)
```

```python
endes2022['act_embara'].value_counts()
```

```python
teens = endes2022['Edad']==1

solo_teens =endes2022[teens]
solo_teens
```

```python
solo_teens = solo_teens.copy()
solo_teens['act_embara'] = solo_teens['act_embara'].replace({3: 'Si', 2: 'No', 1: 'Si', 0: 'No'})
```

```python
si_por_provincia = solo_teens[solo_teens['act_embara'] == 'Si']['IDPROV'].value_counts()

# Obtener el conteo total por provincia
total_por_provincia = solo_teens['IDPROV'].value_counts()

# Combinar las series en un nuevo DataFrame
data_prov = pd.DataFrame({'Si_Count': si_por_provincia, 'Total_Count': total_por_provincia})

# Calcular el porcentaje de 'Si' por provincia
data_prov['Porcentaje'] = (data_prov['Si_Count'] / data_prov['Total_Count']) * 100
data_prov
```

```python
data_prov= data_prov.reset_index(drop=False).\
             rename_axis(index=None, columns=None) # no name for indexes
data_prov
```

```python
data_prov = data_prov.fillna(0)
```

```python
#Data Scaling
columnsToScale=['Porcentaje'] # you can add more columns

from sklearn import preprocessing

minmaxSc = preprocessing.MinMaxScaler(feature_range=(0, 10))

minmaxResult = minmaxSc.fit_transform(data_prov[columnsToScale])

data_prov['Total_minMax_2022']=minmaxResult[:, 0]
data_prov
```

```python
#Empezamos con el Mapa
import geopandas as gpd
import os

Peru=gpd.read_file(os.path.join("mapas", "peruMaps.gpkg"), layer='prov')
Peru["PROVINCIA"] = Peru["PROVINCIA"].str.strip()
Peru
```

```python
theMap=Peru.merge(data_prov, on=["IDPROV"], how="left")
theMap
```

```python
theMap= theMap.fillna(0)
```

```python
filas_filtradas = theMap[theMap['DEPARTAMEN'] == 'MOQUEGUA']
filas_filtradas
```

```python
#Intervalos
import mapclassify 
import numpy as np

np.random.seed(12345) # so we all get the same results!

k=4

# let's try 5 intervals
theVar2022=theMap.Total_minMax_2022
# based on similarity, optimizer
mp52022 = mapclassify.MaxP(theVar2022, k=4) 
```

```python
theMap['Total_mp52022'] = mp52022.yb
```

```python
changes={0:"0_Muy Bajo", 1:"1_Bajo", 2:"2_Alto", 3:"3_Muy Alto"}
theMap[['Total_mp52022']]=theMap.loc[:,['Total_mp52022']].replace(to_replace=changes)
```
